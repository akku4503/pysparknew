```python
import zipfile
import os

# Define the path to the uploaded zip file
zip_file_path = 'csv_files.zip'

# Define the directory to extract the files to
extracted_dir = 'csv_files'

# Create the extraction directory if it doesn't exist
os.makedirs(extracted_dir, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_dir)

print(f"Files extracted to: {extracted_dir}")

# List the extracted files to verify
extracted_files = os.listdir(extracted_dir)
print("Extracted files:", extracted_files)
```

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType

products_schema = StructType([
    StructField("productCode", StringType()),
    StructField("productName", StringType()),
    StructField("productLine", StringType()),
    StructField("productScale", StringType()),
    StructField("productVendor", StringType()),
    StructField("productDescription", StringType()),
    StructField("quantityInStock", IntegerType()),
    StructField("buyPrice", DoubleType()),
    StructField("MSRP", DoubleType())
])

products_df = spark.read.csv("/content/csv_files/csv_files/products.csv", schema=products_schema, header=True)
products_df.write.mode("overwrite").parquet("/content/data/parquet/products")
```

```python
product_summary = products_df.groupBy("productLine").agg(
    count("productCode").alias("totalProducts"),
    avg("MSRP").alias("avgMSRP"),
    max("MSRP").alias("maxMSRP")
)
product_summary.write.mode("overwrite").parquet("/content/output/processed/products_summary.parquet")
```

```python
orderdetails_df = spark.read.csv("/content/csv_files/csv_files/orderdetails.csv", header=True, inferSchema=True)
orders_df = spark.read.csv("/content/csv_files/csv_files/orders.csv", header=True, inferSchema=True)
```

```python
top_products = orderdetails_df.groupBy("productCode").agg(
    sum("quantityOrdered").alias("totalQty")
).join(products_df, "productCode") \
 .orderBy("totalQty", ascending=False).limit(10)
top_products.write.mode("overwrite").parquet("/content/output/processed/top_10_products.parquet")
```

```python
revenue_by_product = orderdetails_df.withColumn("revenue", col("quantityOrdered") * col("priceEach")) \
    .join(products_df, "productCode") \
    .groupBy("productName").agg(sum("revenue").alias("totalRevenue"))
revenue_by_product.write.mode("overwrite").parquet("/content/output/processed/revenue_by_product.parquet")
```

```python
customers_df = spark.read.csv("/content/csv_files/csv_files/customers.csv", header=True, inferSchema=True)
payments_df = spark.read.csv("/content/csv_files/csv_files/payments.csv", header=True, inferSchema=True)
employees_df = spark.read.csv("/content/csv_files/csv_files/employees.csv", header=True, inferSchema=True)
offices_df = spark.read.csv("/content/csv_files/csv_files/offices.csv", header=True, inferSchema=True)
```

```python
revenue_by_country = customers_df.join(payments_df, "customerNumber") \
    .groupBy("country").agg(sum("amount").alias("totalRevenue"))
revenue_by_country.write.mode("overwrite").parquet("/content/output/processed/revenue_by_country.parquet")
```

```python
from pyspark.sql.functions import countDistinct, sum

office_sales = customers_df.join(payments_df, "customerNumber") \
    .join(employees_df, customers_df["salesRepEmployeeNumber"] == employees_df["employeeNumber"]) \
    .join(offices_df, "officeCode") \
    .groupBy(offices_df["officeCode"], offices_df["city"], offices_df["country"]) \
    .agg(countDistinct("customerNumber").alias("customerCount"), sum("amount").alias("totalSales"))
office_sales.write.mode("overwrite").parquet("/content/output/processed/customer_sales_by_office.parquet")
```

```python
employee_summary = employees_df.join(customers_df, employees_df["employeeNumber"] == customers_df["salesRepEmployeeNumber"]) \
    .join(payments_df, "customerNumber") \
    .groupBy("employeeNumber", "firstName", "lastName") \
    .agg(countDistinct("checkNumber").alias("totalOrders"), sum("amount").alias("totalSales"))
employee_summary.write.mode("overwrite").parquet("/content/output/processed/employee_sales_summary.parquet")
```

```python
lazy_df = orders_df.groupBy("status").count()
print("Before Action")
lazy_df.show()
print("After Action")
```

```python
product_demand = orderdetails_df.groupBy("productCode").agg(
    count("orderNumber").alias("orderCount"),
    sum("quantityOrdered").alias("totalOrderedQty"),
    avg("priceEach").alias("averagePriceEach")
).join(products_df, "productCode") \
 .select("productCode", "productName", "orderCount", "totalOrderedQty", "averagePriceEach")

product_demand.write.mode("overwrite").parquet("/content/output/processed/product_demand_summary.parquet")
```
